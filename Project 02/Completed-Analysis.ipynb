2. **Platform with highest engagement**  
   - Which platform had the highest **engagement**?  
   - *Note: Engagement = likes + shares + comments*

```python
import plotly.express as px

engagement_by_platform = db.query("""
    SELECT 
        platform,
        SUM(likes + shares + comments) AS Engagement
    FROM df
    GROUP BY platform
""").df()

# Sort highest → lowest
engagement_by_platform = engagement_by_platform.sort_values("Engagement", ascending=True)

# Highlight highest (last after sorting)
engagement_by_platform["color"] = [
    "darkgray" if i != len(engagement_by_platform) - 1 else "crimson"
    for i in range(len(engagement_by_platform))
]

fig = px.bar(
    engagement_by_platform,
    x="Engagement",
    y="platform",   # 🔥 swapped axes for inversion
    orientation="h",
    color="color",
    color_discrete_map="identity",
    template="simple_white",
    title="Total Engagement by Platform (Inverted Axis View)"
)

# Zoomed scale (x-axis now holds values)
x_min = engagement_by_platform["Engagement"].min()
x_max = engagement_by_platform["Engagement"].max()
padding = (x_max - x_min) * 0.2

fig.update_layout(
    title_x=0,
    showlegend=False,
    xaxis=dict(
        title="Total Engagement",
        range=[x_min - padding, x_max + padding]  # zoomed emphasis
    ),
    yaxis=dict(
        title="Platform",
        categoryorder="total ascending"  # ensures highest appears at top
    )
)

fig.show()
```

### Platform Performance Insights from the Bar Chart

The bar chart indicates that **Instagram** is the highest-performing platform, outperforming **Twitter** by a slim margin. The difference between the two platforms amounts to only a few tens of thousands, highlighting how closely matched their engagement levels are. In contrast, **Facebook** emerges as the lowest-performing platform, with total engagement of approximately **1.32 million**.

To improve visual clarity and make the small marginal differences between platforms more discernible, the **Total_Engagement** axis was shifted upward. This adjustment allows the relative variations in performance to be clearly presented without altering the underlying data.

3. **Average interactions per post**  
   - What was the **average number of likes, shares, comments, and views** per post on each platform?    

```python
df_metrics = db.query("""
    SELECT 
        platform,
        post_type,
        ROUND(AVG(likes), 2) AS avg_likes,
        ROUND(AVG(shares), 2) AS avg_shares,
        ROUND(AVG(comments), 2) AS avg_comments
    FROM df
    GROUP BY platform, post_type
    ORDER BY platform, post_type
""").df()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

platforms = df_metrics["platform"].unique()
metrics = ["avg_likes", "avg_shares", "avg_comments"]

# Create 3 rows (one per metric)
fig = make_subplots(
    rows=len(metrics),
    cols=len(platforms),
    subplot_titles=[
        f"{metric} - {platform}"
        for metric in metrics
        for platform in platforms
    ],
    vertical_spacing=0.08
)

for col_i, platform in enumerate(platforms, start=1):

    df_plat = df_metrics[df_metrics["platform"] == platform]

    for row_i, metric in enumerate(metrics, start=1):

        values = df_plat[["post_type", metric]].copy()

        fig.add_trace(
            go.Bar(
                x=values["post_type"],
                y=values[metric],
                marker_color="lightslategray",  # neutral only
                showlegend=False
            ),
            row=row_i,
            col=col_i
        )

# -------------------------------
# Layout (clean + neutral)
# -------------------------------
fig.update_layout(
    template="simple_white",
    title="Engagement Metrics by Platform (Neutral Comparison View)",
    title_x=0,
    height=800,
    margin=dict(l=60, r=40, t=80, b=40)
)

# Axis labels per metric row
fig.update_yaxes(title_text="Avg Likes", row=1)
fig.update_yaxes(title_text="Avg Shares", row=2)
fig.update_yaxes(title_text="Avg Comments", row=3)

for i in range(1, len(platforms) + 1):
    fig.update_xaxes(title_text="Post Type", row=3, col=i)

fig.show()
```

### Insights from Subplots by Metric and Platform

Subplots were created for each metric (e.g., average likes), segmented by platform and further broken down by the three different post types. Several notable insights emerge from this visualization:

- Across all platforms, **image-based posts consistently achieve relatively higher performance** compared to other post types.
- Engagement levels across platforms and post types appear **stable and closely clustered**, with no extreme spikes or drops.
- The close alignment of values suggests that while post type does influence performance, **differences are marginal rather than drastic**, indicating broadly similar audience engagement behavior across platforms.

Overall, the subplot approach makes it easier to compare post-type performance within and across platforms, revealing consistent patterns rather than sharp contrasts.

   4. **Total number of posts by type**  
      - What was the **total number of each type of post** (video, image, text) on each platform?

```python
post_counts = db.query("""
    SELECT 
        platform,
        post_type,
        COUNT(*) AS Tot_Posts
    FROM df
    GROUP BY platform, post_type
    ORDER BY post_type, Tot_Posts DESC
""").df()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

platforms = post_counts["platform"].unique()

fig = make_subplots(
    rows=1,
    cols=len(platforms),
    subplot_titles=[f"{p}" for p in platforms]
)

for i, platform in enumerate(platforms, start=1):

    df_plat = post_counts[post_counts["platform"] == platform]

    # -------------------------------
    # COLOR STYLE (list-based like example)
    # -------------------------------
    colors = ["lightslategray"] * len(df_plat)

    # highlight ONLY the max post type
    max_idx = df_plat["Tot_Posts"].idxmax()
    highlight_position = df_plat.index.get_loc(max_idx)
    colors[highlight_position] = "crimson"

    # -------------------------------
    # BAR TRACE
    # -------------------------------
    fig.add_trace(
        go.Bar(
            x=df_plat["post_type"],
            y=df_plat["Tot_Posts"],
            marker_color=colors,
            showlegend=False
        ),
        row=1,
        col=i
    )

    # -------------------------------
    # VALUE LABEL (top only)
    # -------------------------------
    top_row = df_plat.loc[max_idx]

    fig.add_annotation(
        x=top_row["post_type"],
        y=top_row["Tot_Posts"],
        text=f"{top_row['Tot_Posts']}",
        showarrow=False,
        yshift=10,
        font=dict(size=11),
        bgcolor="rgba(255,255,255,0.8)",
        row=1,
        col=i
    )

# -------------------------------
# LAYOUT
# -------------------------------
fig.update_layout(
    template="simple_white",
    title="Highest Performing Post Type per Platform",
    title_x=0,
    margin=dict(l=60, r=40, t=80, b=40)
)

fig.update_yaxes(title_text="Total Posts")
fig.update_xaxes(title_text="Post Type")

fig.show()
```

### Post Type Distribution Across Platforms

Using a bar chart, subplots were created for each platform to visualize the total number of posts by post type. The results show that **image posts are the most common post type across all platforms**.

This prevalence may indicate a strategic preference for image-based content, potentially because such posts tend to perform better in terms of engagement. This is especially relevant for a visually driven platform like **Instagram**, where image posts align closely with user expectations and consumption patterns. As a result, the dominance of image posts may reflect both historical performance trends and platform-specific content strategies.

5. **Post type with highest average engagement**  
   - Which **type of post** had the highest **average engagement**?

```python
import plotly.express as px

# -------------------------------
# 1. Query data
# -------------------------------
avg_eng = db.query("""
    SELECT 
        post_type,
        ROUND(AVG(likes + shares + comments), 0) AS avg_Engagement
    FROM df
    GROUP BY post_type
    ORDER BY avg_Engagement DESC, post_type
""").df()

# -------------------------------
# 2. Sort for inverted visual
# -------------------------------
avg_eng = avg_eng.sort_values("avg_Engagement", ascending=True)

# -------------------------------
# 3. Color logic (highlight highest only)
# -------------------------------
avg_eng["color"] = [
    "darkgray" if i != len(avg_eng) - 1 else "crimson"
    for i in range(len(avg_eng))
]

# -------------------------------
# 4. Horizontal bar chart with text labels
# -------------------------------
fig = px.bar(
    avg_eng,
    x="avg_Engagement",
    y="post_type",
    orientation="h",
    color="color",
    color_discrete_map="identity",
    text="avg_Engagement",   # ✅ add value labels
    template="simple_white",
    title="Average Engagement by Post Type (Inverted Axis View)"
)

# -------------------------------
# 5. Axis scaling (zoomed emphasis)
# -------------------------------
x_min = avg_eng["avg_Engagement"].min()
x_max = avg_eng["avg_Engagement"].max()
padding = (x_max - x_min) * 0.2

fig.update_layout(
    title_x=0,
    showlegend=False,
    xaxis=dict(
        title="Average Engagement",
        range=[x_min - padding, x_max + padding]
    ),
    yaxis=dict(
        title="Post Type",
        categoryorder="total ascending"
    )
)

# -------------------------------
# 6. Improve text formatting
# -------------------------------
fig.update_traces(
    texttemplate="%{text:,.0f}",
    textposition="outside"
)

# -------------------------------
# 7. Display
# -------------------------------
fig.show()
```

### Average Engagement by Post Type

The results were plotted using a bar chart to illustrate average engagement by post type across platforms. The visualization indicates that **text-based posts achieve the highest average engagement across all platforms**.

This finding suggests that, despite the widespread use of image posts, text-based content may resonate more strongly with audiences on average. It highlights the importance of content substance and messaging, and indicates that post frequency does not necessarily correlate with higher engagement performance.

6. **Total engagement per quarter**  
   - What was the **total engagement** in each **quarter** of the year?

```python
import plotly.graph_objects as go

# Query quarterly engagement
eng_trend = db.query("""
    SELECT 
        Quarter(date) AS quarters,
        ROUND(SUM(likes + shares + comments), 0) AS Engagement
    FROM df
    GROUP BY quarters
    ORDER BY quarters
""").df()

# -------------------------------
# COLOR LOGIC (list-based style)
# -------------------------------
colors = ["lightslategray"] * len(eng_trend)
max_idx = eng_trend["Engagement"].idxmax()
highlight_position = eng_trend.index.get_loc(max_idx)
colors[highlight_position] = "crimson"

# -------------------------------
# LINE + MARKERS
# -------------------------------
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=eng_trend["quarters"],
    y=eng_trend["Engagement"],
    mode="lines+markers",
    line=dict(color="lightslategray", width=2),
    marker=dict(color=colors, size=10),
    showlegend=False
))

# -------------------------------
# AXIS FORMATTING
# -------------------------------
fig.update_layout(
    title="Engagement Per Quarter",
    title_x=0,
    template="simple_white",
    xaxis=dict(
        tickmode="array",
        tickvals=[1, 2, 3, 4],
        ticktext=["Q1", "Q2", "Q3", "Q4"],
        title="Quarter"
    ),
    yaxis_title="Total Engagement"
)

# -------------------------------
# ANNOTATIONS (elevated + underlined)
# -------------------------------
for _, row in eng_trend.iterrows():
    fig.add_annotation(
        x=row["quarters"],
        y=row["Engagement"],
        text=f"<u>{int(row['Engagement']):,}</u>",
        showarrow=False,
        yshift=26,
        font=dict(size=12, color="black")
    )

fig.show()
```

### Quarterly Trends in Sales and Engagement

The line graph shows that **sales have been declining since the first quarter**, eventually **dropping below 1 million in the third quarter**. This downward trend is followed by a **sharp recovery in the fourth quarter**, where sales rise to **over 1.1 million**, indicating a strong rebound toward the end of the year.

In contrast, **total engagement remains relatively stable across all quarters**. Fluctuations are minimal, with engagement levels ranging between **approximately 950,000 at the lowest point** and **1.2 million at the highest**. This stability suggests that while sales experience notable volatility, audience engagement remains consistently strong throughout the year.
